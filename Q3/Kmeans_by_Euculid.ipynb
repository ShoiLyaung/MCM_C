```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
training_set = pd.read_csv('Wordle.csv')
X = training_set.iloc[:, 5:12].values
X_word = training_set.iloc[:, 2:3].values
```

```python
for n_clusters in range (1,10):
    cluster = KMeans(n_clusters=n_clusters,random_state=0).fit(X)
    centroid=cluster.cluster_centers_
    y_pred = cluster.labels_#
    words=[]
    Y=[1,2,3,4,5,6,7]
    centtrod = cluster.cluster_centers_
    color=['blue','orange','green','purple','brown','pink','gray','olive','cyan']
    fig, axi1=plt.subplots(1)
    for i in range(n_clusters):
        axi1.plot(Y,centroid[i,:],c=color[i])
        words_pred=[]
        words.append(words_pred)
        words_pred.append(X_word[y_pred==i,:])
```

```python
n_clusters=5
cluster = KMeans(n_clusters=n_clusters,random_state=0).fit(X)
centroid=cluster.cluster_centers_
y_pred = cluster.labels_#
Y=[1,2,3,4,5,6,7]
centtrod = cluster.cluster_centers_
color=['blue','orange','green','purple','brown','pink','gray','olive','cyan']
words={}
for i in range(n_clusters):
    fig, axi1=plt.subplots(1)
    axi1.plot(Y,centroid[i,:],c=color[i])
    label_word = X_word[y_pred==i,:]
    words[str(i)] = label_word.reshape(len(label_word)).tolist()
```

```python
import json
words_json = json.dumps(words, sort_keys=False, indent=4, separators=(',', ': '))
with open("labeled_words.json", "w") as f:
    f.write(words_json)
```
