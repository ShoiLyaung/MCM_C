```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Wordle.csv')
words = df.loc[:, ["Word"]].values
words = words.reshape(len(words)).tolist()
df
```

```python
vowels = ['a', 'e', 'i', 'o', 'u']
vowel_num = []
for word in words:
    num = 0
    for letter in word:
        if letter in vowels:
            num = num + 1
    vowel_num.append(num / 5.0)
```

```python
df["Percentage of vowel in a word"] = vowel_num
df_corr = df[["Percentage of hard mode", "Percentage of frequency", "Percentage of vowel in a word"]]
df_corr
```

```python
d = df_corr.corr("spearman")
sns.set_context({'figure.figsize':[7, 7]})
sns.heatmap(d, cmap="YlOrRd", linewidths=0.1, vmax=1.0, square=True, linecolor='white', annot=True)
```
